```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Lotka-Volterra Model

```python
# VARIABLES

 # Constantes 
Cxx = 0.1
Cxy = 0.002
Cyx = 0.0025
Cyy = 0.2


params = [Cxx,Cxy,Cyx,Cyy]

# Delta X
h = 0.1

# The Variable Time (t) set of 0 to 2
tend = 100
ntsteps = int(tend/h)

t = np.linspace(0, tend, ntsteps)
print(t)
```

```python
# INITIAL CONDITIONS 

x0 , y0 = [80.0,20.0]

print(x0 , y0 )

# Euler

x_y = np.zeros([2, ntsteps]) # Array bidimensional de dos dimesiones
print(x_y)

x_y[0, 0] = x0  # Y0 esta es la condicion incial que las colocas dentro del array de dos dimensiones en esa posicion
x_y[1, 0] = y0  # Y1   
print("x_y =\n", x_y)

vel_x_y = np.zeros(ntsteps) #  Array de una dimension para la energia 
print("velocity =\n",vel_x_y)


# Implicit Euler Method

x_y_ieuler = np.zeros([2, ntsteps]) # Array bidimensional de dos dimesiones
print(x_y_ieuler)

x_y_ieuler[0, 0] = x0  # Y0 esta es la condicion incial que las colocas dentro del array de dos domensiones en esa posicion
x_y_ieuler[1, 0] = y0  # Y1   
print("x_y =\n", x_y_ieuler)

vel_x_y_ieuler = np.zeros(ntsteps) #  Array de una dimension para la energia 
print("velocity =\n",vel_x_y_ieuler)


# Verlet

x_y_verlet = np.zeros([2, ntsteps]) # Array bidimensional de dos dimesiones
print(x_y_verlet)

x_y_verlet[0, 0] = x0  # Y0 esta es la condicion incial que las colocas dentro del array de dos domensiones en esa posicion
x_y_verlet[1, 0] = y0  # Y1   
print("x_y =\n", x_y_verlet)

vel_x_y_verlet = np.zeros(ntsteps) #  Array de una dimension para la energia 
print("velocity =\n",vel_x_y_verlet)


# Ruggen Kuta

x_y_RK = np.zeros([2, ntsteps]) # Array bidimensional de dos dimesiones
print(x_y_RK)

x_y_RK[0, 0] = x0  # Y0 esta es la condicion incial que las colocas dentro del array de dos domensiones en esa posicion
x_y_RK[1, 0] = y0  # Y1   
print("x_y =\n", x_y_RK)

vel_x_y_RK = np.zeros(ntsteps) #  Array de una dimension para la energia 
print("velocity =\n",vel_x_y_RK)


# Scipy

x_y_scipy = np.zeros([2, ntsteps]) # Array bidimensional de dos dimesiones
print(x_y_scipy)

x_y_scipy[0, 0] = x0  # Y0 esta es la condicion incial que las colocas dentro del array de dos domensiones en esa posicion
x_y_scipy[1, 0] = y0  # Y1   
print("x_y =\n", x_y_RK)

vel_x_y_scipy = np.zeros(ntsteps) #  Array de una dimension para la energia 
print("velocity =\n",vel_x_y_scipy)

```

```python
# Euler 
def dev(x_y,t):
    return np.array([  params[0] * x_y[0] - params[1] * x_y[0] * x_y[1],
                     - params[3] * x_y[1] + params[2] * x_y[0] * x_y[1]])        

def euler(x_y,t):
    return np.array(x_y + dev(x_y, t) * h)


# Implicit Euler Method

def dydtsym(x_y, t):
    return np.array([  params[0] * x_y[0] - params[1] * x_y[1] *  x_y[0], 
                     - params[3] *  x_y[1] + params[2] * x_y[1] * (x_y[0] + h * (params[0] * x_y[0] - params[1] * x_y[1] *  x_y[0]) )])

def eulersym(x_y, t):
    return np.array(x_y + dydtsym(x_y, t) * h)


# Verlet

def verlet(x_y, t):
    vertelym = x_y[1] + h/2 * (- params[3] * x_y[1] + params[2] * x_y[0] * x_y[1])
    verletx = x_y[0] + h * ( params[0] * x_y[0] - params[1] * x_y[0] * vertelym)
    verlety = vertelym + h/2 * (- params[3] *  vertelym+ params[2] *  verletx *  vertelym)
    return np.array([verletx, verlety])


# Ruggen Kuta

def RK(x_y,t):
    K1 = np.array((  params[0] * x_y[0] - params[1] * x_y[0] * x_y[1], 
                   - params[3] * x_y[1] + params[2] * x_y[0] * x_y[1]))
    
    K2 = np.array(((  params[0] * x_y[0] - params[1] * x_y[0] * (x_y[1] + h * K1[1]/2)),
                   (- params[3] * x_y[1] + params[2] * x_y[1] * (x_y[0] + h * K1[0]/2))))
                  
    K3 = np.array((( params[0] * x_y[0] - params[1] * x_y[0] * (x_y[1] + h * K2[1]/2)),  
                   (- params[3] * x_y[1] + params[2] * x_y[1] * (x_y[0] + h * K2[0]/2))))
                  
    K4 = np.array((  params[0] * x_y[0] - params[1] * x_y[0] * (x_y[1] + h * K3[1]), 
                   - params[3] * x_y[1] + params[2] * x_y[1] * (x_y[0] + h * K3[0])))
    
    Sol = x_y + (h * (K1 + 2* K2 + 2* K3 + K4)/6)
    
    return Sol 

```

```python
# iteration steps
for j in range(ntsteps-1):
    x_y[:, j+1] = euler(x_y[:, j], t[j])
    x_y_ieuler[:, j+1] = eulersym(x_y_ieuler[:, j], t[j])
    x_y_verlet[:, j+1] = verlet(x_y_verlet[:, j], t[j])
    x_y_RK[:, j+1]     = RK(x_y_RK[:, j], t[j])
```

```python
# SCIPY
# Initial conditions 

def Lotka_Volerra_model(x_y):
    x_y[0], x_y[1] = x_y
    x =   params[0] * x_y[0] - params[1] * x_y[0] * x_y[1]   
    y = - params[3] * x_y[1] + params[2] * x_y[0] * x_y[1]
    return np.array([x,y])
```

```python
# SCIPY
# Initial conditions 

x0 , y0 = [80.0,20.0]

Cxx = 0.1
Cxy = 0.002
Cyx = 0.0025
Cyy = 0.2


def Lotka_Volerra_model(x_y, t, Cxx, Cxy,Cyx,Cyy):
    x_y[0], x_y[1] = x_y
    x =  Cxx * x_y[0] - Cxy * x_y[0] * x_y[1]   
    y = - Cyy * x_y[1] + Cyx * x_y[0] * x_y[1]
    return np.array([x,y])


t = np.linspace(0, tend, ntsteps)
X0 = [x0, y0]

res = odeint(Lotka_Volerra_model, X0, t, args = (Cxx, Cxy,Cyx,Cyy))
x, y = res.T

print(res.T)
```

```python
# V = c yx x  c yy ln( x) + c xy y  c xx ln( y)
def velocity(x_y):
    velocity = params[2]*x_y[0] - params[3]*np.log(x_y[0]) + params[1]*x_y[1] - params[0]*np.log(x_y[1])
    return velocity

params = [Cxx,Cxy,Cyx,Cyy]
```
